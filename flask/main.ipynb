```python
import matplotlib.pyplot as plt
import solvers
```

```python
# examples
a = 0
b = 2
T = 1

# k = lambda u, x, t: u**1
q = lambda u, x, t: 1

f = lambda u, x, t: 0

# mu_a = lambda t: 1*t**(-0.2)#1*t**(-0.2)
mu_a = lambda K: [lambda t: 0, lambda t: 1, lambda t: 1*t**(K)]#1*t**(-0.2)]
mu_b = [lambda t: 0, lambda t: 1, lambda t: 0]

u0 = lambda x: 0

tau = 0.01
h = 0.01
```

```python
fig, axs = plt.subplots(3, 2)
fig.set_figheight(15)
fig.set_figwidth(15)

U, Xs, Ts = solvers.ParabolicPartialDifferentialEquation(q, lambda u, x, t: u**1, f, u0, mu_a(0), mu_b, a, b, T, h, tau)
axs[0, 0].plot(Xs, U[20], color="green")
axs[0, 0].plot(Xs, U[40], color="yellow")
axs[0, 0].plot(Xs, U[60], color="orange")
axs[0, 0].plot(Xs, U[80], color="red")
axs[0, 0].plot(Xs, U[-1], color="blue")
axs[0, 0].set_title('n=1, k=0')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('U')

U, Xs, Ts = solvers.ParabolicPartialDifferentialEquation(q, lambda u, x, t: u**2, f, u0, mu_a(0), mu_b, a, b, T, h, tau)
axs[0, 1].plot(Xs, U[20], color="green")
axs[0, 1].plot(Xs, U[40], color="yellow")
axs[0, 1].plot(Xs, U[60], color="orange")
axs[0, 1].plot(Xs, U[80], color="red")
axs[0, 1].plot(Xs, U[-1], color="blue")
axs[0, 1].set_title('n=2, k=0')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('U')

U, Xs, Ts = solvers.ParabolicPartialDifferentialEquation(q, lambda u, x, t: u**1, f, u0, mu_a(1), mu_b, a, b, T, h, tau)
axs[1, 0].plot(Xs, U[20], color="green")
axs[1, 0].plot(Xs, U[40], color="yellow")
axs[1, 0].plot(Xs, U[60], color="orange")
axs[1, 0].plot(Xs, U[80], color="red")
axs[1, 0].plot(Xs, U[-1], color="blue")
axs[1, 0].set_title('n=1, k=1')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('U')

U, Xs, Ts = solvers.ParabolicPartialDifferentialEquation(q, lambda u, x, t: u**2, f, u0, mu_a(1/2), mu_b, a, b, T, h, tau)
axs[1, 1].plot(Xs, U[20], color="green")
axs[1, 1].plot(Xs, U[40], color="yellow")
axs[1, 1].plot(Xs, U[60], color="orange")
axs[1, 1].plot(Xs, U[80], color="red")
axs[1, 1].plot(Xs, U[-1], color="blue")
axs[1, 1].set_title('n=2, k=1/2')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('U')

U, Xs, Ts = solvers.ParabolicPartialDifferentialEquation(q, lambda u, x, t: u**3, f, u0, mu_a(1), mu_b, a, b, T, h, tau)
axs[2, 0].plot(Xs, U[20], color="green")
axs[2, 0].plot(Xs, U[40], color="yellow")
axs[2, 0].plot(Xs, U[60], color="orange")
axs[2, 0].plot(Xs, U[80], color="red")
axs[2, 0].plot(Xs, U[-1], color="blue")
axs[2, 0].set_title('n=3, k=1')
axs[2, 0].set_xlabel('X')
axs[2, 0].set_ylabel('U')

U, Xs, Ts = solvers.ParabolicPartialDifferentialEquation(q, lambda u, x, t: u**1, f, u0, mu_a(-0.2), mu_b, a, b, T, h, tau)
axs[2, 1].plot(Xs, U[20], color="green")
axs[2, 1].plot(Xs, U[40], color="yellow")
axs[2, 1].plot(Xs, U[60], color="orange")
axs[2, 1].plot(Xs, U[80], color="red")
axs[2, 1].plot(Xs, U[-1], color="blue")
axs[2, 1].set_title('n=1, k=-0.2')
axs[2, 1].set_xlabel('X')
axs[2, 1].set_ylabel('U')

plt.show()
```
