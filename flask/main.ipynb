```python
import time
#Numerical method for solving the system of linear equations
#Seidel method

def Seidel(A, b, eps):
    x = x_new = np.zeros(len(A))
    D = np.diag(np.diag(A))
    A1 = np.tril(A)
    while True:
        x_new = x + np.dot(np.linalg.inv(A1+D), b - np.dot(A, x))
        if np.linalg.norm(x_new - x) < eps:
            break
        x = x_new.copy()
    return x

#Without matrix operations
def Seidel2(A, b, eps):
    x = np.zeros(len(A))
    x_new = np.zeros(len(A))
    while True:
        for i in range(len(A)):
            x_new[i] = (b[i] - np.sum(A[i][:i] * x_new[:i]) - np.sum(A[i][i+1:] * x[i+1:])) / A[i][i] 
        if np.linalg.norm(x_new - x) < eps:
            break
        x = x_new.copy()
    return x

#Jacobi method
def Jacobi(A, b, eps):
    x = x_new = np.zeros(len(A))
    D = np.diag(np.diag(A))
    A1 = np.tril(A)
    A2 = np.triu(A)
    while True:
        x_new = x + np.dot(np.linalg.inv(D), b - np.dot(A, x))
        if np.linalg.norm(x_new - x) < eps:
            break
        x = x_new.copy()
    return x

#Without matrix operations
def Jacobi2(A, b, eps):
    x = np.zeros(len(A))
    x_new = np.zeros(len(A))
    while True:
        for i in range(len(A)):
            x_new[i] = (b[i] - sum(A[i][j] * x[j] for j in range(len(A)) if i!=j)) / A[i][i] 
        if np.linalg.norm(x_new - x) < eps:
            break
        x = x_new.copy()
    return x

#Example
#Generate a random symmetric matrix A
NN = 1000
A = np.random.rand(NN, NN)
A = np.dot(A, A.T) + np.eye(NN) * NN*NN
# print(A)
#Generate a random vector b
b = np.random.rand(NN)
#Check matrix compatibility
if np.linalg.det(A) == 0:
    print("System isn't compatible")
else:
    #Machine precision
    eps = np.finfo(float).eps
    # perfomance plot
    start = time.time()
    solution = Seidel(A, b, eps)
    work_time = time.time() - start
    solution = 0
    print("Solution, using Seidel method: ", solution, "Time: ", work_time)
    start = time.time()
    solution = Seidel2(A, b, eps)
    work_time = time.time() - start
    solution = 0
    print("Solution, using Seidel2 method: ", solution, "Time: ", work_time)
    
    start = time.time()
    solution = Jacobi(A, b, eps)
    work_time = time.time() - start
    solution = 0
    print("Solution, using Jacobi method: ", solution, "Time: ", work_time)
    
    start = time.time()
    solution = Jacobi2(A, b, eps)
    work_time = time.time() - start
    solution = 0
    print("Solution, using Jacobi2 method: ", solution, "Time: ", work_time)
    
    #Compare with the result of the numpy.linalg.solve function
    start = time.time()
    solution = np.linalg.solve(A, b)
    work_time = time.time() - start
    solution = 0
    print("Solution, using numpy.linalg.solve: ", solution, "Time: ", work_time)
```

```python
4.645965099334717
#Функция нахождения нормы Фробениуса матрицы
def NormF(A):
    return pow(sum(abs(A[i][j])**2 for i in range(len(A)) for j in range(len(A))), 1/2)

#Нахождение обратной матрицы методом Гаусса
def InverseMatrix(A):
    #Создание единичной матрицы
    E = np.eye(len(A))
    #Создание расширенной матрицы
    A = np.concatenate((A, E), axis=1)
    #Прямой ход
    for i in range(len(A)):
        #Нахождение максимального элемента в столбце
        max_el = abs(A[i][i])
        max_row = i
        for k in range(i + 1, len(A)):
            if abs(A[k][i]) > max_el:
                max_el = abs(A[k][i])
                max_row = k
        #Поменять местами строку с максимальным элементом и текущую
        for k in range(len(A[i])):
            A[max_row][k], A[i][k] = A[i][k], A[max_row][k]
        #Деление строки на максимальный элемент
        for k in range(len(A[i])):
            A[i][k] = A[i][k] / max_el
        #Вычитание строк
        for k in range(len(A)):
            if k != i:
                c = A[k][i] / A[i][i]
                for j in range(len(A[i])):
                    A[k][j] = A[k][j] - c * A[i][j]
    #Создание обратной матрицы
    A = A[:, len(A):]
    return A

#Функция нахождения числа обусловленности матрицы
def ConditionNumber(A, Norm):
    return Norm(A) * Norm(InverseMatrix(A))

print(ConditionNumber(A, NormF))
print(np.linalg.cond(A))
```
