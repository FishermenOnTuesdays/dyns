```python
#Numerical method for solving the system of linear equations
#Seidel method

def Seidel(A, b, eps):
    x = x_new = np.zeros(len(A))
    while True:
        x_new = x + (b - np.dot(A, x)) / np.diag(A)
        if np.linalg.norm(x_new - x) < eps:
            break
        x = x_new.copy()
    return x

#Example
A = np.array([[10, -1, 2, 0], [-1, 11, -1, 3], [2, -1, 10, -1], [0, 3, -1, 8]])
b = np.array([6, 25, -11, 15])
#Machine precision
eps = np.finfo(float).eps
print(Seidel(A, b, eps))

#Compare with the result of the numpy.linalg.solve function
print(np.linalg.solve(A, b))
```
